## Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist

#创建画布
fig = plt.figure(figsize=(10, 6))
#使用axisartist.Subplot方法创建一个绘图区对象ax
ax = axisartist.Subplot(fig, 111)  
#将绘图区对象添加到画布中
fig.add_axes(ax)
x = np.arange(-10, 10, .1)
y = 1 / (1 + np.exp(-x))

#通过set_visible方法设置绘图区所有坐标轴隐藏
ax.axis[:].set_visible(False)

#ax.new_floating_axis代表添加新的坐标轴
ax.axis["x"] = ax.new_floating_axis(0,0)
#给x坐标轴加上箭头
ax.axis["x"].set_axisline_style("-|>", size = 2.0)
#添加y坐标轴，且加上箭头
ax.axis["y"] = ax.new_floating_axis(1,0)
ax.axis["y"].set_axisline_style("-|>", size = 2.0)
#设置x、y轴上刻度显示方向
ax.axis["x"].set_axis_direction("top")
ax.axis["y"].set_axis_direction("right")
ax.axhline(y=1, color='g', ls='--')
ax.plot(x, y)
plt.savefig('source/sigmoid_funtion.svg', transparent=True)
```

## tanh Function

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist

#创建画布
fig = plt.figure(figsize=(10, 6))
#使用axisartist.Subplot方法创建一个绘图区对象ax
ax = axisartist.Subplot(fig, 111)  
#将绘图区对象添加到画布中
fig.add_axes(ax)
x = np.arange(-10, 10, .1)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

#通过set_visible方法设置绘图区所有坐标轴隐藏
ax.axis[:].set_visible(False)

#ax.new_floating_axis代表添加新的坐标轴
ax.axis["x"] = ax.new_floating_axis(0,0)
#给x坐标轴加上箭头
ax.axis["x"].set_axisline_style("-|>", size = 2.0)
#添加y坐标轴，且加上箭头
ax.axis["y"] = ax.new_floating_axis(1,0)
ax.axis["y"].set_axisline_style("-|>", size = 2.0)
#设置x、y轴上刻度显示方向
ax.axis["x"].set_axis_direction("top")
ax.axis["y"].set_axis_direction("right")
ax.axhline(y=-1, color='g', ls='--')
ax.axhline(y=1, color='g', ls='--')
ax.plot(x, y)
plt.savefig('source/tanh_funtion.svg', transparent=True)
```

## ReLU

## Early stopping

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist

#创建画布
fig = plt.figure(figsize=(10, 6))
#使用axisartist.Subplot方法创建一个绘图区对象ax
ax = axisartist.Subplot(fig, 111)  
#将绘图区对象添加到画布中
fig.add_axes(ax)
x = np.arange(0.5, 15, .1)
y1 = 1 / (x*0.1)
y2 = 1 / (x*0.1) + .4*x

#通过set_visible方法设置绘图区所有坐标轴隐藏
ax.axis[:].set_visible(False)

#ax.new_floating_axis代表添加新的坐标轴
ax.axis["x"] = ax.new_floating_axis(0,0)
#给x坐标轴加上箭头
ax.axis["x"].set_axisline_style("-|>", size = 2.0)
ax.set_xlabel("Number of iterations")
#添加y坐标轴，且加上箭头
ax.axis["y"] = ax.new_floating_axis(1,0)
ax.axis["y"].set_axisline_style("-|>", size = 2.0)

#设置x、y轴上刻度显示方向
ax.axis["x"].set_axis_direction("bottom")
ax.axis["y"].set_axis_direction("left")

ax.axis["y"].set_label("Loss")

ax.plot(x, y1, label='train loss')
ax.axvline(x=5, color='r', ls='--')
ax.plot(x, y2, label='dev loss')
ax.legend(loc='upper right', facecolor='None')
plt.savefig('source/early_stopping.svg', transparent=True)
```

## Batch vs Mini-batch

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist

#创建画布
fig = plt.figure(figsize=(14, 6))
#使用axisartist.Subplot方法创建一个绘图区对象ax
ax = axisartist.Subplot(fig, 121)  
ax2 = axisartist.Subplot(fig, 122)
#将绘图区对象添加到画布中
fig.add_axes(ax)
fig.add_axes(ax2)
x = np.arange(4, 200)
y = 10 / np.sqrt(x)
y_ = y*np.exp(np.random.randn(y.shape[0])*0.08)
#通过set_visible方法设置绘图区所有坐标轴隐藏
ax.axis[:].set_visible(False)

#ax.new_floating_axis代表添加新的坐标轴
ax.axis["x"] = ax.new_floating_axis(0,0)
#给x坐标轴加上箭头
ax.axis["x"].set_axisline_style("-|>", size = 2.0)
#添加y坐标轴，且加上箭头
ax.axis["y"] = ax.new_floating_axis(1,0)
ax.axis["y"].set_axisline_style("-|>", size = 2.0)
#设置x、y轴上刻度显示方向
ax.axis["x"].set_axis_direction("bottom")
ax.axis["y"].set_axis_direction("left")
ax.set_ylim(0, 5)
# ax.set_xlim(-1, 200)


ax.axis["x"].set_label("Number of iterations")
ax.axis["y"].set_label("cost")
ax.set_title('Batch gradient descent')
ax.plot(x, y)
#通过set_visible方法设置绘图区所有坐标轴隐藏
ax2.axis[:].set_visible(False)

#ax.new_floating_axis代表添加新的坐标轴
ax2.axis["x"] = ax2.new_floating_axis(0,0)
#给x坐标轴加上箭头
ax2.axis["x"].set_axisline_style("-|>", size = 2.0)
#添加y坐标轴，且加上箭头
ax2.axis["y"] = ax2.new_floating_axis(1,0)
ax2.axis["y"].set_axisline_style("-|>", size = 2.0)
#设置x、y轴上刻度显示方向
ax2.axis["x"].set_axis_direction("bottom")
ax2.axis["y"].set_axis_direction("left")
ax2.axis["x"].set_label("Number of iterations")
ax2.axis["y"].set_label("cost")
ax2.plot(x, y_)
ax2.set_ylim(0, 5)
ax2.set_title('Mini-batch gradient descent')


plt.savefig('source/batch_or_mini-batch.svg', transparent=True)
```
